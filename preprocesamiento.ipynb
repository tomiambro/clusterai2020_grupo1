```python
# importamos las librerías necesarias para trabajar.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#Otros 
import warnings
warnings.filterwarnings('ignore')
```

### Descargar dataset crudo de:

https://datos.gob.ar/dataset/salud-covid-19-casos-registrados-republica-argentina

```python
#abro los datasets
completo = pd.read_csv("./data/Covid19Casos.csv", delimiter=',', parse_dates=['fecha_apertura','fecha_diagnostico','fecha_cui_intensivo','fecha_internacion','fecha_fallecimiento','fecha_inicio_sintomas'])   
completo.head()
```

```python
df_caba = completo.loc[completo['residencia_provincia_nombre'] == 'CABA', :]
df_caba.fecha_apertura.sort_values().tail()
```

```python
df_caba = df_caba.loc[df_caba.fecha_apertura <= pd.Timestamp(year=2020, month=9, day=30), :]
df_caba.fecha_apertura.sort_values().tail()
```

```python
df_caba = df_caba.drop('ultima_actualizacion', axis=1)
```

```python
df_caba.reset_index().to_feather('./data/CovidCabaFeather')
```
